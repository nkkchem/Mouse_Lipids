```python
def extract_table_from_excel(filename, sheet_name=0, newcol='Uniprot', col='RazorProtein', delimiter='|', extract_col=1):
    df = pd.read_excel(filename,sheet_name=0)
    df[newcol] = df[col].str.split(delimiter).str.get(extract_col)
    return df.set_index(newcol)

lcm4 = extract_table_from_excel('LCM4_MaxQuant_Protein_Normalized_Data.xlsx')
lcm3 = extract_table_from_excel('LCM3_MaxQuant_Protein_Normalized_Data.xlsx')
display(lcm3), display(lcm4)
```

```python
lmpd = pd.read_table('LMPD_040215 2.txt')
mouse_lmpd = lmpd[lmpd['species'] == 'Mouse'].set_index('uniprot_id')

```
